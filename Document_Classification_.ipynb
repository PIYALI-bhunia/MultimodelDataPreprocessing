```python
import numpy as np
import pandas as pd
import pathlib
```

```python
data_root = pathlib.Path('/content/drive/MyDrive/tobaco_OCR/')

for item in data_root.iterdir():
  print(item)
```

```python
def get_file_paths_and_labels(data_root):
  text_paths = [str(path) for path in data_root.glob('*/*.txt')]
  return text_paths
```

```python
text_paths= get_file_paths_and_labels(data_root)
text_paths
```

```python
texts = []
for text_path in text_paths:
  with open(text_path, 'r') as f:
    texts.append(f.read())
```

```python
df = pd.DataFrame(list(zip(text_paths, texts)),
               columns =['path', 'text'])
df
```

```python
import re
```

```python
# removing punctuations
df['text']=[re.sub(r'[^\w\s\n]','',data) for data in df['text']]
```

```python
df = df.replace('\n','', regex=True)
```

```python
df
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
df['text'] = df['text'].apply(nltk.word_tokenize)
```

```python
df['text']= df['text'].apply(lambda x: [i.lower() for i in x ] )
df
```

```python
#stopwords remove
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words
```

```python
def remove_stopwords(text):
    new_wordlist= [i for i in text if i not in stop_words]
    return new_wordlist
```

```python
df['text']= df['text'].apply(lambda x:remove_stopwords(x))
df
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer  = WordNetLemmatizer()
lemmatized_dataset = []

for this_str in df['text']:
    lemmatized_str = []
    for word in this_str:
        lemmatized_str.append(lemmatizer.lemmatize(word))
    lemmatized_dataset.append(lemmatized_str)

df['text'] = lemmatized_dataset
df
```
