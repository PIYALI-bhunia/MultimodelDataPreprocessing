# Text vectorization using OneHotEncoding

```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/COVID-19 AR Clinical Correlates July202020_1.csv")
print(dataset.shape)
dataset.head()
```

```python
dataset.columns=dataset.iloc[0]
dataset=dataset[1:]
dataset
```

```python
dataset=dataset[['patient_id','Radiologist KEY FINDINGS']]
dataset
```

```python
dataset.dropna(inplace=True)
dataset.shape
```

```python
dataset['Radiologist KEY FINDINGS'] = dataset.groupby(['patient_id'])['Radiologist KEY FINDINGS'].transform(lambda x : ' '.join(x))
  
# drop duplicate data
dataset = dataset.drop_duplicates()
dataset
```

```python
import re
```

```python
# removing punctuations
dataset['Radiologist KEY FINDINGS']=[re.sub(r'[^\w\s]','',data) for data in dataset['Radiologist KEY FINDINGS']]
```

```python
dataset
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
#Tokenization
dataset['Radiologist KEY FINDINGS'] = dataset['Radiologist KEY FINDINGS'].apply(nltk.word_tokenize)
```

```python
dataset
```

```python
dataset['Radiologist KEY FINDINGS']= dataset['Radiologist KEY FINDINGS'].apply(lambda x: [i.lower() for i in x ] )
```

```python
dataset
```

```python
#stopwords remove
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_wordlist= [i for i in text if i not in stop_words]
    return new_wordlist
```

```python
dataset['Radiologist KEY FINDINGS']= dataset['Radiologist KEY FINDINGS'].apply(lambda x:remove_stopwords(x))
```

```python
dataset
```

```python
# Lemmatization
from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer  = WordNetLemmatizer()
lemmatized_dataset = []

for this_str in dataset['Radiologist KEY FINDINGS']:
    lemmatized_str = []
    for word in this_str:
        lemmatized_str.append(lemmatizer.lemmatize(word))
    lemmatized_dataset.append(lemmatized_str)

dataset['Radiologist KEY FINDINGS'] = lemmatized_dataset
dataset
```

```python
bigger_list=list(dataset['Radiologist KEY FINDINGS'])
print(bigger_list)
```

```python
#find unique word in the bigger list
unique_word=[]
for i in bigger_list:
  for j in i:
    if j not in unique_word:
      unique_word.append(j)
print(len(unique_word))
```

```python
#find maximum length of array in bigger list
MAX=0
for i in bigger_list:
  if len(i)>MAX:
    MAX=len(i)
print(MAX)
```

```python
token_index = {}
counter = 0

for i in bigger_list:                                      
  for considered_word in i:
    if considered_word not in token_index:
      token_index.update({considered_word : counter + 1}) 
      counter+= 1
```

```python
print(token_index)
```

```python
max_length = 172
results  = np.zeros(shape = (len(bigger_list),
                            max_length,
                            max(token_index.values()) + 1))  
```

```python
for i, sample in enumerate(bigger_list): 
  for j, considered_word in list(enumerate(sample)):
    index = token_index.get(considered_word)
    results[i, j, index] = 1.                 
```

```python
results
```

```python
s1,s2,s3=results.shape
```

```python
results=results.reshape(s1, s2*s3)
results.shape
```

```python

df1=pd.DataFrame(results)
```

```python
df1
```

```python
filepath =  r"/content/drive/MyDrive/COVID-19 AR Clinical Correlates July202020.csv"
df2 = pd.read_csv(filepath, header = 1)
df2.tail()
```

```python
mortality_truth = list(df2['MORTALITY'])[:-2]
mortality_truth=[0 if i =='N' else 1 for i in mortality_truth]
df1['MORTALITY'] = mortality_truth
final_dataset=df1
final_dataset
```

```python
X =final_dataset.iloc[:,:-1]
y =final_dataset['MORTALITY']

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 5)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
```

```python
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
