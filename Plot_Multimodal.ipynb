## TEXT DATA

```python
import pandas as pd
import numpy as np

data={'RandomForest':[59.97,74.46,0,0],'SVM':[59.54,48.92,0,0],'ANN':[58.53,62.87,0,0],'CNN_1D':[0,0,76.61,79.05]}
df=pd.DataFrame(data,columns=['RandomForest','SVM','ANN','CNN_1D'],index=['BERT','TFIDF','FastText','Glove'])

ax = df.plot(kind='bar',figsize=(20, 10),rot=0)
 
ax.set_title('models vs their accuracy in % only for text data')
ax.set_xlabel("models implemented")
ax.set_ylabel("Accuracy")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

##Image Data

```python
import pandas as pd
import numpy as np

data={'Accuracy':[77.2,78.54,80.65],'Validation_Accuracy':[75.53,78.78,81.64]}
df=pd.DataFrame(data,columns=['Accuracy','Validation_Accuracy'],index=['MobileNetV2','Resnet50','VGG16'])

ax = df.plot(kind='bar',figsize=(10, 10),rot=0)
ax.set_title('models vs their accuracy in % only for image data')
ax.set_xlabel("models implemented for image data")
ax.set_ylabel("Accuracy")
ax.legend(loc='upper left', fontsize=9, bbox_to_anchor=(1,1)) 
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

## Multimodal Data

```python
import pandas as pd
import numpy as np

data={'Accuracy':[83.62,85.34],'Validation_Accuracy':[85.1,85.96]}
df=pd.DataFrame(data,columns=['Accuracy','Validation_Accuracy'],index=['Bert_MobileNetV2','Bert_VGG16'])

ax = df.plot(kind='bar',figsize=(8,6),rot=0) 
ax.set_title('models vs their accuracy in % only for multimodal data')
ax.set_xlabel("models implemented for both image and text data")
ax.set_ylabel("Accuracy")
ax.legend(loc='center left', fontsize=9, bbox_to_anchor=(1,1))
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
rf_accuracy = 74.46
svm_accuracy = 48.92
ann_accuracy = 62.87
  
# creating the dataset
data = {'Random Forest': rf_accuracy, 'SVM':svm_accuracy, 'ANN':ann_accuracy}


models = list(data.keys())
accuracy = list(data.values())
  
fig = plt.figure(figsize = (8, 4))
 
# creating the bar plot
plt.bar(models, accuracy , color ='maroon',
        width = 0.3)
 
plt.xlabel("models implemented")
plt.ylabel("accuracy")
plt.title("TFIDF - word embedding, (only Text modality)")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

models = ['random forest', 'SVM','ANN']
accuracy = [74.46, 48.92,62.87]
 
# Figure Size
fig, ax = plt.subplots(figsize =(10, 5))
 
# Horizontal Bar Plot
ax.barh(models, accuracy)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('models vs their accuracy in %',
             loc ='left', )
# Show Plot
plt.show()
```
