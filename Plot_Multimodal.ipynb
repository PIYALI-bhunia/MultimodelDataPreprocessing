## TEXT DATA

```python
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
w=0.2
x=["TF_IDF","BERT","FastText","Glove","ELMO"]
fig = plt.figure(figsize = (15, 10))
RandomForest=[74.46,59.97,70,80,40]
SVM=[48.92,59.54,43,80,70]
ANN=[62.87,58.53,50,70,40]
CNN_1D=[50,70,76.61,79.05,50]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
bar3=[i+w for i in bar2]
bar4=[i+w for i in bar3]
plt.bar(bar1,RandomForest,w,label="RandomForest")
plt.bar(bar2,SVM,w,label="SVM")
plt.bar(bar3,ANN,w,label="ANN")
plt.bar(bar4,CNN_1D,w,label="CNN_1D")

plt.xlabel("models implemented")
plt.ylabel("accuracy")
plt.title('models vs their accuracy in % only for text data')
plt.xticks(bar1+w,x)
plt.legend(loc='upper left')
plt.show()
```

##Image Data

```python
w=0.2
x=["MobileNetV2","Resnet_50","VGG_16"]
fig = plt.figure(figsize = (8, 6))
Accuracy=[77.2,67.82,80.65]
Val_Accuracy=[75.53,65.33,81.64]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
plt.bar(bar1,Accuracy,w,label="Accuracy")
plt.bar(bar2,Val_Accuracy,w,label="Validation Accuracy")

plt.xlabel("models implemented")
plt.ylabel("accuracy")
plt.title('models vs their accuracy in % only for image data')
plt.xticks(bar1+w/2,x)
plt.legend(loc='upper center')
plt.show()
```

## Multimodal Data

```python
w=0.2
x=["BERT_MobileNetV2","BERT_Resnet50","BERT_VGG16","Glove_MobileNetV2","Glove_Resnet50","Glove_VGG16"]
fig = plt.figure(figsize = (10, 8))
Accuracy=[77.2,67.82,80.65,0,0,0]
Val_Accuracy=[75.53,65.33,81.64,0,0,0]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
plt.bar(bar1,Accuracy,w,label="Accuracy")
plt.bar(bar2,Val_Accuracy,w,label="Validation Accuracy")

plt.xlabel("Multimodal implemented")
plt.ylabel("accuracy")
plt.title('models vs their accuracy in % only for both image and text data')
plt.xticks(bar1+w/2,x)
plt.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd

data={'RandomForest':[74.46,59.97,80,78,90],'SVM':[48.92,59.54,20,0,0],'ANN':[62.87,58.53,50,60,70],'CNN_1D':[0,0,0,0,0]}
df=pd.DataFrame(data,columns=['RandomForest','SVM','ANN','CNN_1D'],index=['TF_IDF','BERT','FastText','Glove','ELMO'])

ax = df.plot(kind='bar',figsize=(15,10),width=1.005) 

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(10, 5))
 
# Horizontal Bar Plot
ax.barh(data={'RandomForest':[74.46,59.97,60,50],'SVM':[48.92,59.54,40,30],'ANN':[62.87,58.53,50,70]},index=["TF_IDF","BERT","FastText","Glove","ELMO"])
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('models vs their accuracy in %',
             loc ='left', )
# Show Plot
plt.show()
 
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
rf_accuracy = 74.46
svm_accuracy = 48.92
ann_accuracy = 62.87
  
# creating the dataset
data = {'Random Forest': rf_accuracy, 'SVM':svm_accuracy, 'ANN':ann_accuracy}


models = list(data.keys())
accuracy = list(data.values())
  
fig = plt.figure(figsize = (8, 4))
 
# creating the bar plot
plt.bar(models, accuracy , color ='maroon',
        width = 0.3)
 
plt.xlabel("models implemented")
plt.ylabel("accuracy")
plt.title("TFIDF - word embedding, (only Text modality)")
plt.show()
```
